```python
def desaninhar_colunas(df):
    while True:
        novas_colunas = []
        for col in df.columns:
            # Se tiver dicionário
            if df[col].apply(lambda x: isinstance(x, dict)).any():
                df_norm = pd.json_normalize(df[col])
                df_norm.columns = [f"{col}_{sub}" for sub in df_norm.columns]
                df = pd.concat([df.drop(columns=[col]), df_norm], axis=1)
                break  # recomeça o loop, pois df mudou

            # Se tiver lista de dicionários
            elif df[col].apply(lambda x: isinstance(x, list) and all(isinstance(i, dict) for i in x if i is not None)).any():
                df = df.explode(col, ignore_index=True)
                df_norm = pd.json_normalize(df[col])
                df_norm.columns = [f"{col}_{sub}" for sub in df_norm.columns]
                df = pd.concat([df.drop(columns=[col]), df_norm], axis=1)
                break  # recomeça o loop, pois df mudou
        else:
            break  # nenhuma coluna restante para processar
    return df

```

```python
df_candidatos = desaninhar_colunas(candidatos)
```

```python
df_vagas = desaninhar_colunas(vagas)
```

```python
df_prospect = desaninhar_colunas(prospect)
```
