### Nesse projeto utilizamos um modelo de classificação Kneighbors para prever se a pessoa precisará de entubação baseado nos sintomas e comorbidades
conseguimos uma precisão de 78%

```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Definindo a URL do dataset
url = "https://datalakedtat.blob.core.windows.net/techchallenge3/pnad_covid_mes_9_10_11.microdados2.csv"
```

```python
# Lendo o dataset
df = pd.read_csv(url, sep=',')
```

```python
# Exibindo as primeiras linhas do dataset
df.head()
```

```python
# Utilizando o LabelEncoder para transformar os valores da coluna 'teste_positivo' em valores numéricos
encoder = LabelEncoder()
encoder.fit(df['teste_positivo'])
df['teste_positivo'] = encoder.transform(df['teste_positivo'])
```

```python
# Utilizando o LabelEncoder para transformar os valores da coluna 'uf' em valores numéricos
encoder.fit(df['uf'])
df['uf'] = encoder.transform(df['uf'])
```

```python
# Criando um lista com as colunas que serão utilizadas para a predição
df_corr = df.loc[:, ['febre', 'dor_olhos', 'cheiro_sabor', 'dor_muscular', 'diarreia',
       'tosse', 'dor_garganta', 'dificuldade_respiracao', 'dor_cabeca',
       'dor_peito', 'nausea', 'nariz_ntupido', 'fadiga', 'diabetes', 'hipertesao',
       'doenca_resp', 'doenca_coracao', 'depressao', 'cancer', 'teste_positivo',
       'entubado_resp_art', 'idade']]
```

```python
# Criando um gráfico de correlação
corr = df_corr.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# Criando um dataframe com as colunas utilizadas para a predição
df_ml = df_corr.loc[:, ['febre', 'dor_olhos', 'cheiro_sabor', 'dor_muscular', 'diarreia', 'tosse', 'dor_garganta', 'dificuldade_respiracao', 'dor_cabeca', 'dor_peito', 'nausea', 'nariz_ntupido', 'fadiga', 'diabetes', 'hipertesao', 'doenca_resp', 'doenca_coracao', 'depressao', 'cancer', 'entubado_resp_art', 'idade', 'teste_positivo']]
df_ml = df_ml.dropna()
```

```python
# Definindo a variavel entuda_resp_art como inteiro
df_ml['entubado_resp_art'] = df_ml['entubado_resp_art'].astype(int)
```

```python
# Eliminando valores iguais a 9
df_ml = df_ml[df_ml.entubado_resp_art != 9]
```

```python
# Dividindo o dataset em variáveis dependentes e independentes
y = df_ml.loc[:, 'entubado_resp_art']
x = df_ml.loc[: ,[ 'febre', 'dor_olhos', 'cheiro_sabor', 'dor_muscular', 'diarreia', 'tosse', 'dor_garganta', 'dificuldade_respiracao', 'dor_cabeca', 'dor_peito', 'nausea', 'nariz_ntupido', 'fadiga', 'diabetes', 'hipertesao', 'doenca_resp', 'doenca_coracao', 'depressao', 'cancer', 'idade']]
```

```python
# Dividindo o dataset em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Treinado o modelo
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train, y_train)
```

```python
# Criando a previsão
y_pred_k = knn.predict(x_test)
```

```python
# Testando a acurácia do modelo
acuracia_k = accuracy_score(y_test, y_pred_k)
print(acuracia_k)
```
